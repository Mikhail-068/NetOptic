# Парсинг

## Парсинг единчных тегов

Функция парсинга тега в единичном экземпляре. Например **model, vendor, price**

```python
def parse_one_tag(offers, tag):

  # создаём пустой список
  result = []

  # для каждого товара
  for offer in offers:

    # ищем тег и сохраняем в переменную
    text = offer.find(tag)

    # если тег есть
    if text is not None:

      # если текст в этом теге это пустая строка, то добавляем в список None
      if text.text == '':
        result.append(None)

      else:
        # иначе текст под этим тегом добавляем в список
        result.append(text.text)

    else:
      # если такого тега не нашлось, то добавляем None
      result.append(None)

  # возвращаем итоговый список
  return result
```

функция отображения информации о списке

```python
def show_info(list_):
  print('Всего значений в списке: {}'.format(len(list_)))
  print('Значений None: {}'.format(list_.count(None)))
  print(list_[:10])
```

```python
# список наименований
names_list = parse_one_tag(offers, 'name')

show_info(names_list)
```

```python
# список ссылок
urls_list = parse_one_tag(offers, 'url')

show_info(urls_list)
```

```python
# список цен
prices_list = parse_one_tag(offers, 'price')

show_info(prices_list)
```

```python
# список брендов
vendors_list = parse_one_tag(offers, 'vendor')

show_info(vendors_list)
```

код ниже переводит названия брендов в Camel Case (каждой слово с большой буквы)

```python
# итерируем по длинне списка брендов
for i in range(len(vendors_list)):

  # разделяем по пробелу и полученный список сохраняем в переменную
  new_value = vendors_list[i].split()

  # для каждого элемента списка применяем .capitalize и соединяем в единую строку
  new_value = ' '.join(j.capitalize() for j in new_value)

  # перезаписываем название бренда на новое
  vendors_list[i] = new_value

show_info(vendors_list)
```

```python
# список моделей
models_list = parse_one_tag(offers, 'model')

show_info(models_list)
```

код ниже выделяет артикул/номенклатуру модели там где это возможно

```python
# итерируемся по длинне списка моделей
for i in range(len(models_list)):

  # если название бренда встречается в названии модели
  if vendors_list[i].lower() in models_list[i].lower():

    # делаем сплит по запятой
    new_value = models_list[i].split(',')

    # если после сплита больше одного значения
    if len(new_value) > 1:

      # соединяем артикул обратно начиная со второго элемента списка
      new_value = ','.join(j for j in new_value[1:]).strip()

      # перезаписываем значение на новое
      models_list[i] = new_value

show_info(models_list)
```

## Парсинг нескольких одинаковых тегов

функция парсинга тегов в нескольких экземплярах, в данном случае это только ссылки на изображения

```python
def parse_some_tags(offers, tag):

  # это будет список списков, в каждом подсписке ссылки на изображения
  result = []

  # для каждого товара
  for offer in offers:

    # создаём список для ссылок на изображения
    list_ = []

    # для каждого изображения
    for i in offer.findAll(tag):

      # если под тегом не пусто и он начинается с 'http'
      if i and i.text.startswith('http'):

        # в список добавляем ссылку
        list_.append(i.text)

    # если лист со ссылками не пустой, то добавляем его в итоговый список
    if list_:
      result.append(list_)

    # если никаких ссылок на изображения нет, то добавляем None
    else:
      result.append(None)

  # возвращаем список
  return result
```

```python
# список ссылок на изображения
pictures_list = parse_some_tags(offers, 'picture')

show_info(pictures_list)
```

```python
# создаём словарь для будущего датафрейма, под ключами сохраняем соответствующие списки
data_dict = {}
data_dict['Бренд'] = vendors_list
data_dict['Наименование'] = names_list
data_dict['Цена'] = prices_list
data_dict['Ссылка'] = urls_list
data_dict['Модель'] = models_list
data_dict['Изображения'] = pictures_list
```

## Парсинг именованных тегов

функция парсинга именованных тегов, в данном случае тег **param** содержит разные атрибуты очков: **param name="Категория"**, **param name="Ширина оправы (мм)"** и так далее

```python
def parse_named_tag(offers, name):

  # создаём пустой список
  result = []

  # для каждого товара
  for offer in offers:

    # ищем тег с определённым именем атрибута
    text = offer.find('param', {'name': name})

    # если найдено
    if text is not None:
      # добавляем значение в список
      result.append(text.text.lower())

    # иначе добавляем None
    else:
      result.append(None)

  # возвращаем результат
  return result
```

Я решил выделить все теги **param** и их названия сохранить в список

```python
# создаём пустой список
named_params = []

# для каждого товара
for offer in offers:

  # ищем все теги 'param' и проходим циклом каждый
  for param in offer.findAll('param'):
    # сохраняем то, что содержится в name='' в переменную
    name = list(param.attrs.values())[0]

    # если такого параметра ещё нет в списке, то добавляем его
    if name not in named_params:
      named_params.append(name)

# выведем все именованные параметры
print(named_params)
```

код ниже проходит по всем тегам **param** из списка выше

```python
# для каждого именованного атрибута
for param in named_params:

  # выводим какой атрибут обрабатываем
  print('\nОбработка параметра: {}'.format(param))

  # в словарь по ключу - названию атрибута - сохраняем данные, которые возвращает функция
  data_dict[param] = parse_named_tag(offers, param)

  # отображаем информацию по каждому атрибуту после парсинга
  show_info(data_dict[param])
```
